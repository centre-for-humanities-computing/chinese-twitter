```python
# Outputs a df that has id, created_at, hashtag - hashtags are clean, date is normalized to datetime
def hashtag_df_long(large_df, remove_hashtags = False):
    # Get the tokens and dates
    data_tokens = pd.DataFrame([dat for dat in large_df.tokens])
    data_dates = pd.DataFrame([dat for dat in large_df.created_at])
    data_sentiment = pd.DataFrame([dat for dat in large_df.sentiment])

    # Join them to a dataframe
    data_tokens["created_at"] = data_dates
    data_tokens['id'] = range(0, len(data_tokens))
    data_sentiment['id'] = range(0, len(data_sentiment))
    
    import re
    # Some lines have ,#
    exp = re.compile('\,#')
    # Some don't have the comma
    exp2 = re.compile('#')
    d = "#"

    # Extract the hashtags
    # Empty list for [id] [DATE],[HASHTAGS] 
    hash_list3 = [[], [],[]]
    # Loop over each tweet in data
    for tweet in range(0,len(data_tokens["created_at"])):
        for key in data_tokens.ner[tweet]:
            # If the key stands for a hashtag, append the hashtag and date to the list
            if data_tokens.ner[tweet][key] == "HASHTAG": #lst[(np.char.mod('%d', j))]]
                # If the tweet is for some reason something like this: "#PreRolledPapers,#herbStash,#420life,#stoner"
                if re.search(exp, data_tokens.token[tweet][key]):
                    # Break it into pieces and append accordingly
                    tinys = data_tokens.token[tweet][key].split(",")
                    for el in tinys:
                        hash_list3[0].append(data_tokens.id[tweet])
                        hash_list3[2].append(el)
                        hash_list3[1].append(data_tokens.created_at[tweet])
                elif re.search(exp2, data_tokens.token[tweet][key]):
                    # Break it into pieces and append accordingly
                    smalls = [d+e for e in data_tokens.token[tweet][key].split(d) if e]
                    for fl in smalls:
                        hash_list3[0].append(data_tokens.id[tweet])
                        hash_list3[2].append(fl)
                        hash_list3[1].append(data_tokens.created_at[tweet])
                else:
                    hash_list3[0].append(data_tokens.id[tweet])
                    hash_list3[2].append(data_tokens.token[tweet][key])
                    hash_list3[1].append(data_tokens.created_at[tweet])
    
    dd = pd.DataFrame(zip(hash_list3[0], hash_list3[1], hash_list3[2]), columns = ["id", "created_at", "hashtag"])

    # Lowercase the hashtags
    dd['hashtag'] = dd.hashtag.str.lower()

    if remove_hashtags:
        # Remove the meh hashtags
        rmv_hashtags = ["#covid19dk", "#dkpol", "#dkmedier", "#coronadk", "#coronavirus", "#covid19", "#coronavirusdk", 
                            "#covid2019dk", "#covid_19", "#covidãƒ¼19", "#covid", "#covid2019", "#coronavirusoutbreak", "#corona",
                           "#2019coronavirus", "#2019n_cov", "#2019ncov", "#covid19denmark", "#corona19dk", "#covid-19", "#ncov2019",
                       "#covid_2019", "#ncov", "#virus", "#covid-19dk", "#å¨ç¥žv"]
        dd = dd[~dd['hashtag'].isin(rmv_hashtags)]
    
    # Cleaning hashtags
    # Remove the hashtags with â€¦
    dd = dd[dd['hashtag'].str.contains('â€¦$')==False]
    # Remove hashtags that contain .
    dd = dd[dd['hashtag'].str.contains('\.')==False]
    # Remove hashtags that have https in them:
    dd = dd[dd['hashtag'].str.contains('http')==False]
    
    # Add sentiment score based on id
    dd = pd.merge(dd, data_sentiment, on="id")
    
    return(dd)
```

## Converting time bins to source-target (for HoloViews visuals)

```python
def f(x):
  if x['label'] == 'NEGATIVE': return -x['score']
  elif x['label'] == 'POSITIVE': return x['score']
  else: return 0
```

```python
# Convert timebins to source-target
def timebin(timebin_dataframe, with_count_1 = True):
    # Aggregate the dataframe based on ID
    hashtags_per_tweet = timebin_dataframe.groupby(['id'])["hashtag"].apply(lambda x: ', '.join(x)).reset_index()

    cols = [[],[]]
    for index, row in hashtags_per_tweet.iterrows():
        hashtags=row['hashtag'].split()
        hashtags_len = len(hashtags)
        for n in list(itertools.combinations(hashtags, 2)):
            cols[0].append(n[0])
            cols[1].append(n[1])

    # Make the data frame from the source and target columns, remove commas
    df3 = pd.DataFrame(list(zip(cols[0], cols[1]))).rename(columns={0:"source",1:"target"}).replace(',','', regex=True)
    
    # Aggregate repeated edges
    df10 = pd.DataFrame({'count' : df3.groupby(['source', 'target']).size()}).reset_index()
    
    if with_count_1 == False:
        df10 = df10[df10["count"] != 1]
    
    # Add the frequencies of hashtags per whole data + per timebin
    time1 = timebin_dataframe.rename(columns={'hashtag':'source', 'whole_freq': 'source_whole_freq', 'bin_freq':'source_bin_freq'})
    time1 = time1.drop(columns=['id', 'created_at'])

    time2 = timebin_dataframe.rename(columns={'hashtag':'target', 'whole_freq': 'target_whole_freq', 'bin_freq':'target_bin_freq'})
    time2 = time2.drop(columns=['id', 'created_at'])#, 'label', 'score'])

    df11 = pd.merge(df10, time1, how = 'left', on='source')
    df11 = pd.merge(df11, time2, how = 'left', on='target')
    
    # Drop duplicates if any!
    df11 = df11.drop_duplicates().reset_index(drop=True)
    
    return(df11)
```

```python
# Let's take a subset of necessary columns, add id
df = large_df[["created_at", "users"]].reset_index().rename(columns={"index": "id"})
df = data[["created_at", "users"]].reset_index().rename(columns={"index": "id"})
df

# Select only the ones where we have more than 1 hashtag per tweet
df = df[df["users"].map(len) > 1].reset_index(drop=True)

# Hashtag per row
# convert list of pd.Series then stack it
dd = (df
 .set_index(['created_at','id'])['users']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_2', axis=1)
 .rename(columns={0:'user_solo'}))
#lowercase!
dd["user_solo"] = dd["user_solo"].str.lower()

# Fix dates
# Reading the data out and in again to fix datetime
#dd.to_csv("co_hashtag_timebins/own_hashtags_large_hashtags_all.csv", index=False)
#dd = pd.read_csv("co_hashtag_timebins/own_hashtags_large_hashtags_all.csv")

dd.to_csv("co_hashtag_timebins/chinese_users.csv", index=False)
dd = pd.read_csv("co_hashtag_timebins/chinese_users.csv")

# Fix datetime - for some reason doesn't give NaT's only when you read the data in from csv
# Normalize the dates
def clean_dates(dd):
    # 0[DAY]  1[MONTH]  2[YEAR]
    datelist = [[],[],[]]
    for line in dd.created_at:
        #If string starts with letters (Mon Mar)
        if re.match(r'[A-Za-z]', line): #Tue Mar 03 20:44:19 +0000 2020
            # Day
            datelist[0].append(re.split(r'(?:[A-Za-z]{3}\s)([A-Za-z]{3})(\s\d{2})(?:.{16})(\d{4})', line)[2])
            # Month - turn month name into month number
            datelist[1].append(datetime.datetime.strptime((re.split(r'(?:[A-Za-z]{3}\s)([A-Za-z]{3})(\s\d{2})(?:.{16})(\d{4})', line)[1]), "%b").month)
            #datelist[1].append("03")
            # Year
            datelist[2].append(re.split(r'(?:[A-Za-z]{3}\s)([A-Za-z]{3})(\s\d{2})(?:.{16})(\d{4})', line)[3])
        else: #2020-02-28
            # Day
            datelist[0].append(re.split(r'(2020)-(\d*)-(\d*)', line)[3])
            # Month
            datelist[1].append(re.split(r'(2020)-(\d*)-(\d*)', line)[2])
            # Year
            datelist[2].append(re.split(r'(2020)-(\d*)-(\d*)', line)[1])

    # Change the list into DF of datetime
    d = pd.to_datetime(pd.DataFrame(datelist).transpose().rename(columns={0:"day", 1:"month", 2:"year"}))
    dd["created_at"] = d
    return(dd)

dd = clean_dates(dd)
```

```python
def df_to_timebins(id_hashtag):
    
    # Add freq of hashtags by themselves in the dataset
    hashtag_freq = pd.DataFrame({'whole_freq' : id_hashtag.groupby(['user_solo']).size()}).reset_index()

    # Add the whole_frew to id_hashtag - THIS SEEMS BETTER THAN mm
    freq_hashtags = pd.merge(id_hashtag, hashtag_freq, how='left', on=['user_solo'])#, 'id', 'created_at'])
    
    df0 = freq_hashtags
    
    return(freq_hashtags)

f = df_to_timebins(dd)

start_date = "2020-05-20"
end_date = "2020-05-31"
mask = (f['created_at'] >= start_date) & (f['created_at'] <= end_date)
bin0 = f.loc[mask].reset_index(drop=True)

start_date = "2020-06-01"
end_date = "2020-06-30"
mask = (f['created_at'] >= start_date) & (f['created_at'] <= end_date)
bin1 = f.loc[mask].reset_index(drop=True)

start_date = "2020-07-01"
end_date = "2020-07-31"
mask = (f['created_at'] >= start_date) & (f['created_at'] <= end_date)
bin2 = f.loc[mask].reset_index(drop=True)

start_date = "2020-08-01"
end_date = "2020-08-05"
mask = (f['created_at'] >= start_date) & (f['created_at'] <= end_date)
bin3 = f.loc[mask].reset_index(drop=True)

# Convert timebins to source-target
def timebin(timebin_dataframe, with_count_1 = True):
    # Aggregate the dataframe based on ID
    hashtags_per_tweet = timebin_dataframe.groupby(['id'])["user_solo"].apply(lambda x: ', '.join(x)).reset_index()

    cols = [[],[]]
    for index, row in hashtags_per_tweet.iterrows():
        hashtags=row['user_solo'].split()
        hashtags_len = len(hashtags)
        for n in list(itertools.combinations(hashtags, 2)):
            cols[0].append(n[0])
            cols[1].append(n[1])

    # Make the data frame from the source and target columns, remove commas
    df3 = pd.DataFrame(list(zip(cols[0], cols[1]))).rename(columns={0:"source",1:"target"}).replace(',','', regex=True)
    
    # Aggregate repeated edges
    df10 = pd.DataFrame({'count' : df3.groupby(['source', 'target']).size()}).reset_index()
    
    if with_count_1 == False:
        df10 = df10[df10["count"] != 1]
    
    # Add the frequencies of hashtags per whole data + per timebin
    time1 = timebin_dataframe.rename(columns={'user_solo':'source', 'whole_freq': 'source_whole_freq', 'bin_freq':'source_bin_freq'})
    time1 = time1.drop(columns=['id', 'created_at'])

    time2 = timebin_dataframe.rename(columns={'user_solo':'target', 'whole_freq': 'target_whole_freq', 'bin_freq':'target_bin_freq'})
    time2 = time2.drop(columns=['id', 'created_at'])#, 'label', 'score'])

    df11 = pd.merge(df10, time1, how = 'left', on='source')
    df11 = pd.merge(df11, time2, how = 'left', on='target')
    
    # Drop duplicates if any!
    df11 = df11.drop_duplicates().reset_index(drop=True)
    
    return(df11)

# Make the timebins
time0 = timebin(bin0)
time1 = timebin(bin1)
time2 = timebin(bin2)
time3 = timebin(bin3)
#time4 = timebin(f)
```
